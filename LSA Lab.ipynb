```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.datasets import fetch_20newsgroups
import re
import pandas as pd
```

##LSA on the sci.space newsgroup

```python
cat = ['sci.space']
dataset = fetch_20newsgroups(subset='all',shuffle=True, random_state=42, categories=cat)
corpus = dataset.data
```

```python
nltk.download('stopwords')
```

```python
corpus[0:1]
```

####Set all letters to lower case

```python
corpus = [x.lower() for x in corpus]
```

```python
corpus[3]
```

####Removing email addresses

```python
corpus = [re.sub(r'(\s)(\S+\@\S+)(\s)', r'\1\3', corpus[x]) for x in range(len(corpus))]
```

```python
corpus[3]
```

####Adding stopset for common words
#####Update counter: 8

```python
stopset = set(stopwords.words('english'))
stopset.update(['\n', '--', '+', '.', '|', '20', 'would', 'edu', 
                'com', 'net', 'eng', 'umd', 'sysmgr', 'ti', 'uiuc', 
                'ti', 'msu', 'jpl', 'vnet', 'ibm', 're', '_', 'hst', 
                'subject', '__', '___', '____', '22', '575', '3539', 
                'nntp', 'posting', 'host', '10', '30', '50', '14', 
               'henry', 'spencer', 'pat', 'digex', 'access'])
```

####Using TFIDF and LSA code from:
 https://github.com/mbernico/CS570/blob/master/LSA%20Text.ipynb

```python
vectorizer = TfidfVectorizer(stop_words=stopset,
                                 use_idf=True, ngram_range=(1, 3))
X = vectorizer.fit_transform(corpus)
```

```python
X[0]
```

```python
print X[0]
```

```python
X.shape
```

####Using n_components=100 per http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html

```python
lsa = TruncatedSVD(n_components=100, n_iter=100)
lsa.fit(X)
```

```python
lsa.components_[0]
```

###Reviewing a few concepts:
  0. An article written about NASA that discusses space and the moon
  1. Mission/Orbit, and Space station/Shuttle
  2. Space data billion
  3. An article written about rocket elements
  4. Nasa university research
  5. Moon Earth Mars

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_): 
    termsInComp = zip (terms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:10]
    print "Concept %d:" % i
    for term in sortedTerms:
        print term[0]
    print " "
```

```python
lsa.components_
```

####Trying with an increased n_grams range - 2-5 - to see what phrases are commonly combined

```python
vectorizer2 = TfidfVectorizer(stop_words=stopset,
                                 use_idf=True, ngram_range=(2, 5))
X2 = vectorizer2.fit_transform(corpus)
```

```python
lsa2 = TruncatedSVD(n_components=100, n_iter=100)
lsa2.fit(X2)
```
